Soal 2. Selection

Tampilkan data keseluruhan untuk tahun 2015 di bulan November

```python
#code here
import datetime
import pandas as pd

df = pd.read_csv('climate_change.csv')

df['date'] = pd.to_datetime(df['date'])
# print(df.dtypes)

df_nov_2015 = df[
    (df['date'].dt.year == 2015) &
    (df['date'].dt.month == 11)
]

print(df_nov_2015)
```

**Soal 4. Time grouping**

 

*   Lakukan grouping data perubahan iklim pada tahun 2016 berdasarkan frekuensi 1 bulan dengan memberikan agregasi max pada kolom meantemp. humidity, wind_speed dan meanpressure!





```python
#code here

result1 = (
    df[df['date'].dt.year == 2016]
    .groupby(df['date'].dt.to_period('M'))
    [['meantemp', 'humidity', 'wind_speed', 'meanpressure']]
    .max()
)

print (result1)
```

**Soal 5. Time Differenced**

Tunjukkanlah perubahan iklim di tahun 2016 pada minggu terakhir

```python
#code here
year_2016 = df[df['date'].dt.year == 2016]
last_week = year_2016.sort_values('date').tail(7)
result_diff= last_week[["meantemp","humidity","wind_speed","meanpressure"]].diff()
print (result_diff)
```
