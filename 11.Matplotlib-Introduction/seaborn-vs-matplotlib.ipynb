<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# membuat objek figure
fig = plt.figure()

# membuat objek axes
ax = fig.add_subplot()

# menampilkan objek visual
plt.show()
```

```python
#Explicit

# membuat objek figure
fig = plt.figure()

# membuat objek axes
ax = fig.add_subplot()

# menambahkan dalam ke objek axes
ax.plot(x, y)

# menampilkan data
plt.show()
```

```python
#Implicit

# membuat objek visual
plt.plot(x, y)

# menampilkan objek visual
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# aktifkan seaborn style
sns.set()

# plot
plt.plot(x, y, label='cos(x/2)')
plt.plot(x, y2, label='sin(x)')
plt.plot(x, y3, label='sin(2x)')

# tambahan biar rapi
plt.legend()
plt.title("Plot Fungsi Trigonometri")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
